```python
# import the necessary packages
import numpy as np
import pandas as pd
import plotly.express as px

# read in the csv 
df = pd.read_csv('data/dataset.csv')

# remove all columns  except distance and date 

df1 = df[['Date','Distance (m)']]

# drop empty ones 

df1 = df1.dropna()
df1['Date'] = pd.to_datetime(df1['Date'])
display(df1)
```

```python


fig = px.bar(df1, x='Date', y='Distance (m)',
             title='Distance Traveled in bar form')
fig.show()


fig.write_image("./fig1.svg")

```

![](./fig1.svg)

```python
fig1 = px.scatter(df1, x='Date', y='Distance (m)', size='Distance (m)',
                  title='Distance Traveled in scatter form', color='Distance (m)')
fig1.show()

fig1.write_image("./fig2.svg")
```
